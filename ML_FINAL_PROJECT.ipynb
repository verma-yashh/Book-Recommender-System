## Collaborative Filtering System 

```python
# let's create a new dataframe with the ratings with the merging the ratings and the books dataframe

ratings_with_name = ratings.merge(books, on = "ISBN")
```

```python
ratings_with_name.head(5)
```

```python
ratings_with_name.shape
```

```python
ratings_with_name.tail(5)
```

```python
# Lets see the number of ratings with the users 

ratings_with_name.groupby("User-ID").count()["Book-Rating"]
```

```python
# So lets filter out the users that have given more than 200 reviews to counts as well experienced users 

x = ratings_with_name.groupby("User-ID").count()["Book-Rating"] > 200
```

```python
# Using boolean indexing to filter out the users 

x[x].index 
```

```python
# lets store the indexes of the users with more than 100 rartings in a new dataframe 

good_users = x[x].index
```

```python
good_users.shape
```

```python
good_users
```

```python
# Now we will filter out our ratings dataframe with the indexes that we just created 

ratings_with_name["User-ID"].isin(good_users)
```

```python
# Boolean Indexing 

ratings_with_name[ratings_with_name["User-ID"].isin(good_users)]
```

```python
filtered_ratings = ratings_with_name[ratings_with_name["User-ID"].isin(good_users)]
```

```python
y = filtered_ratings.groupby("Book-Title").count()["Book-Rating"] >= 50
```

```python
y
```

```python
popular_books = y[y].index
```

```python
filtered_ratings["Book-Title"].isin(popular_books)
```

```python
final_ratings = filtered_ratings[filtered_ratings["Book-Title"].isin(popular_books)]
```

```python
final_ratings
```

```python
# Lets create a table that has the user indexes with the rating to use the cosine similarity for measuring the distances 

table = final_ratings.pivot_table(index = "Book-Title", columns = "User-ID", values = "Book-Rating")
```

```python
table 
```

```python
table.fillna(0, inplace=True)
```

```python
table
```

```python
# Importing the necesaary library 
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity_score = cosine_similarity(table)
```

```python
similarity_score
```

```python
# Defining the final recommending function 

def recommend(book_name):
    index = np.where(table.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(table.index[i[0]])
```

```python
recommend("1984")
```
